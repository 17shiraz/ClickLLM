```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.font_manager import FontProperties

# === Configuration ===
FONT_PATH = "times.ttf"  # Ensure this file is uploaded in your Colab session
FONT_SIZE = 40
FONT_PROP = FontProperties(fname=FONT_PATH, size=FONT_SIZE)

# === Input confusion matrices ===
confusion_data = {
    "Title": pd.DataFrame(
        [[27516, 2554, 2480], [4843, 1955, 2998], [3019, 1224, 2636]],
        index=["0", "1", "2"], columns=["0", "1", "2"]
    ),
    "Abstract": pd.DataFrame(
        [[23584, 5577, 3389], [2408, 2927, 4461], [1382, 1887, 3610]],
        index=["0", "1", "2"], columns=["0", "1", "2"]
    )
}

# === Plotting function ===
def plot_confusion_matrices(cm_dict, font_prop):
    for label, cm in cm_dict.items():
        # === Plot Graded Confusion Matrix ===
        plt.figure(figsize=(10,8))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                    cbar=False, annot_kws={"fontproperties": font_prop})
        #plt.title(f"{label} - Graded", fontproperties=font_prop)
        plt.xlabel("Predicted", fontproperties=font_prop)
        plt.ylabel("True", fontproperties=font_prop)
        plt.xticks(fontproperties=font_prop)
        plt.yticks(fontproperties=font_prop)
        plt.tight_layout()
        plt.show()

        # === Build Binarized Confusion Matrix ===
        bin_cm = pd.DataFrame([[0, 0], [0, 0]], index=["0", "1"], columns=["0", "1"])
        for i in range(3):
            for j in range(3):
                pred_bin = "0" if i == 0 else "1"
                true_bin = "0" if j == 0 else "1"
                bin_cm.loc[pred_bin, true_bin] += cm.iloc[i, j]

        # === Plot Binarized Confusion Matrix ===
        plt.figure(figsize=(10, 8))
        sns.heatmap(bin_cm, annot=True, fmt="d", cmap="Blues",
                    cbar=False, annot_kws={"fontproperties": font_prop})
        #plt.title(f"{label} - Binary", fontproperties=font_prop)
        plt.xlabel("Predicted", fontproperties=font_prop)
        plt.ylabel("True", fontproperties=font_prop)
        plt.xticks(fontproperties=font_prop)
        plt.yticks(fontproperties=font_prop)
        plt.tight_layout()
        plt.show()

# === Run plotting ===
plot_confusion_matrices(confusion_data, FONT_PROP)
```
